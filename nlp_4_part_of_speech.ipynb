```python
import nltk
from nltk.tag import pos_tag
```

```python
dataset = "1987 – The antiretroviral drug zidovudine (chemical structure pictured) became the first treatment approved by the U.S. Food and Drug Administration for HIV/AIDS."
```

```python
from nltk.tokenize import word_tokenize
```

```python
nltk.download("punkt")
```

```python
nltk.download('punkt_tab')
```

```python
words = word_tokenize(dataset)
```

```python
words
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
pos_tag(words)
```
