```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import joblib
```

```python
x,y = make_classification(n_samples=1000, n_features=4, n_informative=3, n_redundant=1, random_state=42 , n_classes=2)
```

```python
x
```

```python
y
```

```python
df = pd.DataFrame(x,columns=["hours studies","mock_test","attendence","practice_score"])
```

```python
df
```

```python
y
```

```python
df["target"] = y
```

```python
df
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[["hours studies","mock_test","attendence","practice_score"]],df["target"],test_size=0.2,random_state=42)
```

```python
model = Pipeline(steps=[('scaler', StandardScaler()), ('logistic_regression', LogisticRegression(solver='lbfgs',max_iter=1000,verbose=0))])
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
accuracy_score(y_test,model.predict(x_test))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(y_test,model.predict(x_test))
```

```python
print(classification_report(y_test,model.predict(x_test)))
```
