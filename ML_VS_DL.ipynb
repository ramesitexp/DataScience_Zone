```python
import numpy as np
```

```python
x = np.array([[1],[2],[3],[4],[5]])
```

```python
y = np.array([[2],[4],[6],[8],[10]])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ml = LinearRegression()
```

```python
ml.fit(x,y)
```

```python
prediction = ml.predict([[16]])
```

```python
prediction
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
dl = Sequential()
```

```python
dl.add(Dense(units=1,input_shape=(1,)))
```

```python
dl.compile(optimizer='adam',loss='mean_squared_error')
```

```python
dl.fit(x,y,epochs=200)
```

```python
dl.predict(np.array([[10.0]]))
```
