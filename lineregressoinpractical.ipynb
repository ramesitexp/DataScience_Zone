```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import joblib
```

```python
n = 500
area = np.random.randint(600,3500,size=n)
bedroom = np.random.randint(1,6,size=n)
price_lakhs = .035*area + .5*bedroom + np.random.randint(10,50,size=n)
```

```python
df = pd.DataFrame({ 'area': area, 'bedroom': bedroom, 'price_lakhs': price_lakhs })
```

```python
df.head()
```

```python
x = df[['area','bedroom']]
y = df['price_lakhs']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.predict([[6500,3]])
```

```python
r2_score(y_test, model.predict(x_test))
```

```python
joblib.dump(model, 'linear_regression_model.pkl')
```
