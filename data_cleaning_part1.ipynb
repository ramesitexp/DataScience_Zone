```python
dt = df.copy()
cols = ['Gender', 'Area', 'Loan_Status']
```

```python
dt.mode()
```

```python
dt[cols] = dt[cols].fillna(dt.mode().iloc[0])
```

```python
cols2 = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']
```

```python
dt[cols2] = dt[cols2].fillna(dt[cols2].mean())
 # min , max , mean
```
