```python
import numpy as np
import pandas as pd

np.random.seed(42)

n = 400

data = pd.DataFrame({
    "area_sqft": np.random.randint(500, 3000, n),
    "bedrooms": np.random.randint(1, 6, n),
    "bathrooms": np.random.randint(1, 4, n),
    "distance_city_km": np.random.uniform(1, 30, n),
    "age_house": np.random.randint(0, 30, n)
})

# Price formula (realistic logic)
data["price"] = (
    data["area_sqft"] * 6000 +
    data["bedrooms"] * 300000 +
    data["bathrooms"] * 200000 -
    data["distance_city_km"] * 50000 -
    data["age_house"] * 40000 +
    np.random.normal(0, 300000, n)  # noise
)

data.head()
```

```python
from sklearn.model_selection import train_test_split
x = data.drop('price', axis=1)
Y = data['price']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, Y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae =   mean_absolute_error(Y_test, rf.predict(X_test))
rmse =   np.sqrt(mean_squared_error(Y_test, rf.predict(X_test)))
r2 =   r2_score(Y_test, rf.predict(X_test))
print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R^2 Score: {r2}")
```
